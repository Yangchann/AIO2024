```python
import pandas as pd
import numpy as np

data = pd.read_csv('advertising.csv')

def correlation(x, y):
    x_mean = x.mean()
    y_mean = y.mean()
    x_std = x.std()
    y_std = y.std()
    n = len(x)
    return sum((x - x_mean) * (y - y_mean)) / (n * x_std * y_std)

x = data['TV']
y = data['Radio']
corr_xy = correlation(x, y)
print('Correlation between TV and Radio:', round(corr_xy, 2))
```

```python
features = ['TV', 'Radio', 'Newspaper']

for feature1 in features:
    for feature2 in features:
        correlation_value = correlation(data[feature1], data[feature2])
        print(f'Correlation between {feature1} and {feature2}: {round(correlation_value, 2)}')
```

```python
X = data['Radio']
Y = data['Newspaper']

result = np.corrcoef(X, Y)
print(result)
```
