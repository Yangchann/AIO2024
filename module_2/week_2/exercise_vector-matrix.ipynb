#### a. Độ dài vector

```python
import  numpy as np

def compute_vector_length(vector):
    len_of_vector = np.linalg.norm(vector)
    return len_of_vector

vector = np.array([1,2,3,4,5])
print(compute_vector_length(vector))
```

```python
import numpy as np

def compute_vector_length(vector):
    len_of_vector = np.sqrt(np.sum([v**2 for v in vector]))
    return len_of_vector

vector = np.array([1,2,3,4,5])
print(compute_vector_length(vector))
```

#### b. Phép tích vô hướng  

```python
import numpy as np

def compute_dot_product(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    return dot_product

vector1 = np.array([1,2,3,4,5])
vector2 = np.array([5,4,3,2,1])
compute_dot_product(vector1, vector2)
```

```python
import numpy as np

def compute_dot_product(vector1, vector2):
    dot_product = np.sum([v1*v2 for v1,v2 in zip(vector1, vector2)])
    return dot_product

vector1 = np.array([1,2,3,4,5])
vector2 = np.array([5,4,3,2,1])
compute_dot_product(vector1, vector2)
```

#### e. Ma trận nghịch đảo

Cho ma trận  $A = \begin{pmatrix} -2 & 6 \\ 8 & -4 \end{pmatrix}$

Ta được $Det(A)$ = $(-2) * (-4) - 6 * 8 = -40$ 

Vì $Det(A) \neq 0$  nên ma trận $A$ khả nghịch

Suy ra ma trận nghịch đảo của ma trận A:

$A^{-1} = \frac{1}{\text{Det}(A)} \begin{pmatrix} -4 & -6 \\ -8 & -2 \end{pmatrix} = \frac{1}{-40} \begin{pmatrix} -4 & -6 \\ -8 & -2 \end{pmatrix}= \begin{pmatrix} \frac{1}{10} & \frac{3}{20} \\ \frac{1}{5} & \frac{1}{20} \end{pmatrix}$


```python
def inverse_matrix(matrix):
    result = np.linalg.inv(matrix)
    return result
```

```python
matrix  = np.array([[-2,6], [8,-4]])
inverse_matrix(matrix)
```

### 2. Eigenvector và eigenvalues:
---
**Eigenvector và eigenvalues**
- $ A \in \mathbb{R}^{n \times n} $, $ I $ (identity matrix) $ \in \mathbb{R}^{n \times n} $, $ \mathbf{v} \in \mathbb{R}^n $
- Eigenvalue ($ \lambda $): $ \det(A - \lambda I) = 0 $
- Eigenvector ($ \mathbf{v} $): $ A \mathbf{v} = \lambda \mathbf{v} \iff (A - \lambda I) \mathbf{v} = 0 $
- Normalize vector: $ \frac{\mathbf{v}}{\|\mathbf{v}\|} $, $ v_i = \frac{v_i}{\sqrt{\sum_{i=1}^n v_i^2}} $

---

**Cho trước** $ A = \begin{pmatrix} 0.9 & 0.2 \\ 0.1 & 0.8 \end{pmatrix} $

Tìm Eigenvector ($ \mathbf{v} $) đã được normalize và eigenvalue $ \lambda $ của $ A $

---

### **Solution**

Để tìm giá trị riêng, ta giải phương trình đặc trưng $ \det(A - \lambda I) = 0 $:

$ A - \lambda I = \begin{pmatrix} 0.9 - \lambda & 0.2 \\ 0.1 & 0.8 - \lambda \end{pmatrix} $

$ \det(A - \lambda I) = (0.9 - \lambda)(0.8 - \lambda) - (0.2 \times 0.1) = 0 $

Giải định thức:

$(0.9 - \lambda)(0.8 - \lambda) - 0.02 = 0$

$0.72 - 0.9\lambda - 0.8\lambda + \lambda^2 - 0.02 = 0$

$\lambda^2 - 1.7\lambda + 0.7 = 0$

Giải phương trình bậc hai bằng công thức nghiệm:

$\lambda = \frac{1.7 \pm \sqrt{1.7^2 - 4 \times 1 \times 0.7}}{2}$

$\lambda = \frac{1.7 \pm \sqrt{2.89 - 2.8}}{2}$

$\lambda = \frac{1.7 \pm \sqrt{0.09}}{2}$

$\lambda = \frac{1.7 \pm 0.3}{2}$

$\lambda_1 = 1, \quad \lambda_2 = 0.7$


Với mỗi giá trị riêng $ \lambda $, ta tìm vector riêng tương ứng bằng cách giải $ (A - \lambda I)\mathbf{v} = 0 $.

- Với $ \lambda = 1 $:

$A - I = \begin{pmatrix} -0.1 & 0.2 \\ 0.1 & -0.2 \end{pmatrix}$

Giải hệ phương trình:

$\begin{pmatrix} -0.1 & 0.2 \\ 0.1 & -0.2 \end{pmatrix} \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Hệ phương trình này tương đương với:

$-0.1v_1 + 0.2v_2 = 0 \quad \text{và} \quad 0.1v_1 - 0.2v_2 = 0$

Giải hệ, ta được:

$v_2 = \frac{1}{2}v_1$

Vậy, vector riêng là:

$\mathbf{v}_1 = \begin{pmatrix} 1 \\ \frac{1}{2} \end{pmatrix}$

Chuẩn hóa vector riêng:

$\mathbf{v}_1 = \begin{pmatrix} \frac{2}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix}$

- Với $ \lambda = 0.7 $:

$A - 0.7I = \begin{pmatrix} 0.2 & 0.2 \\ 0.1 & 0.1 \end{pmatrix}$

Giải hệ phương trình:

$\begin{pmatrix} 0.2 & 0.2 \\ 0.1 & 0.1 \end{pmatrix} \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Hệ phương trình này tương đương với:

$0.2v_1 + 0.2v_2 = 0 \quad \text{và} \quad 0.1v_1 + 0.1v_2 = 0$

Giải hệ, ta được:

$v_1 = -v_2$

Vậy, vector riêng là:

$\mathbf{v}_2 = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$

Chuẩn hóa vector riêng:

$\mathbf{v}_2 = \begin{pmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{pmatrix}$

---

Giá trị riêng là $ \lambda_1 = 1 $ và $ \lambda_2 = 0.7 $.

Các vector riêng tương ứng (đã chuẩn hóa) là:

$\mathbf{v}_1 = \begin{pmatrix} \frac{2}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix}, \quad \mathbf{v}_2 = \begin{pmatrix} \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} \end{pmatrix}$


```python
import numpy as np

def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

matrix = np.array([[0.9, 0.2], [0.1, 0.8]])
compute_eigenvalues_eigenvectors(matrix)
```

#### 3. Cosine Similarity

* Data (vector $\mathbf{x}, \mathbf{y}$): $\mathbf{x} = \{x_1, \ldots, x_N\}$ $\mathbf{y} = \{y_1, \ldots, y_N\}$


* Cosine Similarity: $\text{cs}(\mathbf{x}, \mathbf{y}) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|} = \frac{\sum_{i=1}^n x_i y_i}{\sqrt{\sum_{i=1}^n x_i^2} \sqrt{\sum_{i=1}^n y_i^2}}$

**Cho** $ \mathbf{x} = \begin{bmatrix} 1 \\ 2 \\ 3 \\ 4 \end{bmatrix} \quad \mathbf{y} = \begin{bmatrix} 1 \\ 0 \\ 3 \\ 0 \end{bmatrix}$

Tìm Cosine similarity $\text{cs}(\mathbf{x}, \mathbf{y})$ 

---
$\textbf{Lời Giải}$

* Tính tích vô hướng của x và y:

$\mathbf{x} \cdot \mathbf{y} = 1 \cdot 1 + 2 \cdot 0 + 3 \cdot 3 + 4 \cdot 0 = 1 + 0 + 9 + 0 = 10$

* Tính độ lớn của từng vector:

$\|\mathbf{x}\| = \sqrt{1^2 + 2^2 + 3^2 + 4^2} = \sqrt{1 + 4 + 9 + 16} = \sqrt{30}$

$\|\mathbf{y}\| = \sqrt{1^2 + 0^2 + 3^2 + 0^2} = \sqrt{1 + 0 + 9 + 0} = \sqrt{10}$

Tính cosine similarity:

$\text{cs}(\mathbf{x}, \mathbf{y}) = \frac{10}{\sqrt{30} \cdot \sqrt{10}} = \frac{10}{\sqrt{300}} = \frac{10}{10\sqrt{3}} = \frac{1}{\sqrt{3}} = \frac{\sqrt{3}}{3}$

Vậy cosine similarity $\text{cs}(\mathbf{x}, \mathbf{y}) = \frac{\sqrt{3}}{3}$.


```python
import numpy as np

def compute_cosine(v1, v2):
    cos_sim = np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
    return cos_sim

x = [1,2,3,4]
y = [1,0,3,0]
compute_cosine(x, y)
```

#### 4. Background subtraction

$\textbf{Resize image and convert color}$

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

IMAGE_SIZE = (678, 381)

bg1_image = cv2.imread("GreenBackground.png", 1)
bg1_image = cv2.resize(bg1_image, IMAGE_SIZE)
bg1_image = cv2.cvtColor(bg1_image, cv2.COLOR_BGR2RGB)

bg2_image = cv2.imread("NewBackground.jpg", 1)
bg2_image = cv2.resize(bg2_image, IMAGE_SIZE)
bg2_image = cv2.cvtColor(bg2_image, cv2.COLOR_BGR2RGB)

ob_image = cv2.imread("Object.png", 1)
ob_image = cv2.resize(ob_image, IMAGE_SIZE)
ob_image = cv2.cvtColor(ob_image, cv2.COLOR_BGR2RGB)
```

$ \textbf{Create function compute\_difference()} $

```python
def compute_difference(image1, image2):
    difference_three_channel = cv2.absdiff(image1, image2)
    difference_single_channel = np.sum(difference_three_channel, axis=2) / 3.0
    difference_single_channel = difference_single_channel.astype(np.uint8)
    return difference_single_channel
```

```python
difference_single_channel = compute_difference(bg1_image, ob_image)
plt.imshow(difference_single_channel)
```

$\text{Create function compute\_binary\_mask()}$

```python
def compute_binary_mask(difference_single_channel):
    difference_binary = np.where(difference_single_channel > 15, 255, 0)
    difference_binary = np.stack([difference_binary]*3, axis=2)
    return difference_binary
```

```python
binary_mask = compute_binary_mask(difference_single_channel)
plt.imshow(binary_mask)
```

$\textbf{Create function replace\_background()}$

```python
def replace_backgrouond(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image, ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)
    output = np.where(binary_mask == 0, bg2_image, ob_image)
    return output
```

```python
background_subtraction = replace_backgrouond(bg1_image, bg2_image, ob_image)
plt.imshow(background_subtraction)
```
