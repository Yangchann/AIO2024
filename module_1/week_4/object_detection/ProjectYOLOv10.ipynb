3. **Training**

```python
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 64

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            imgsz=IMG_SIZE,
            batch=BATCH_SIZE)
```

```python
%pwd
```

4. **Evaluate model**

```python
TRAINED_MODEL_PATH = 'yolov10/runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```

```python
import cv2
from ultralytics import YOLOv10
```

```python
TRAINED_MODEL_PATH = 'yolov10/runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)
conf_thres = 0.5

img = cv2.imread('safety_helmet_dataset/test/images/helmet-10-_jpg.rf.eee985865a1a52407316f0c9ea7a9b3e.jpg')
results = model(img, conf_thres)

for result in results.pred:
    x1, y1, x2, y2 = result[0]
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

cv2.imshow(img)
```
